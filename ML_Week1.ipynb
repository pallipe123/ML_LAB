**Compute Central Tendency Measures: Mean, Median, Mode Measure of Dispersion: Variance, Standard Deviation with different sequences of data having real / integer /categorical /skewed. Also try with data having outliers.**

**1.Numerical Data**

```python
import numpy as np
from statistics import mean, median, mode

data = [10, 12, 14, 15, 18, 20, 22, 25, 30, 35]

print("Data:", data)
print("Mean:", mean(data))
print("Median:", median(data))
print("Mode:", mode(data))
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))
```

**2.Categorical Data**

```python
from statistics import mode

data = ["red", "blue", "red", "green", "red", "blue"]

print("Categorical Data:", data)
print("Mode:", mode(data))
```

**3.Skewed Data**

```python
from statistics import mean, median

data = [2, 3, 4, 5, 6, 100]

print("Skewed Data:", data)
print("Mean:", mean(data))
print("Median:", median(data))
```

**4.Data with Outliers**

```python
from statistics import mean, median

data = [10, 11, 12, 13, 14, 15, 200]

print("Data with Outliers:", data)
print("Mean:", mean(data))
print("Median:", median(data))
```
