```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("synthetic_dataset.csv")

print(df.head())
print(df.info())
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])
print("Selected Numeric Attributes:")
print(numeric_df.columns)
```

```python
corr = numeric_df.corr()
print(corr)
```

```python
# Numerical columns â†’ Mean
for col in df.select_dtypes(include=['int64','float64']).columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Categorical columns â†’ Mode
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Missing values after handling:")
print(df.isnull().sum())
```

```python
num_col = numeric_df.columns[0]

df[num_col + "_Category"] = pd.cut(
    df[num_col],
    bins=3,
    labels=['Low','Medium','High']
)

print(df[[num_col, num_col + "_Category"]].head())
```

```python
df.loc[0, num_col] = df[num_col].mean() + 6 * df[num_col].std()
df.loc[1, num_col] = df[num_col].mean() - 6 * df[num_col].std()
```

```python
Q1 = df[num_col].quantile(0.25)
Q3 = df[num_col].quantile(0.75)
IQR = Q3 - Q1

df = df[(df[num_col] >= Q1 - 1.5*IQR) &
        (df[num_col] <= Q3 + 1.5*IQR)]

print("Outliers removed using IQR method")
```

```python
plt.hist(df[num_col], bins=20)
plt.title("Univariate Analysis")
plt.xlabel(num_col)
plt.ylabel("Frequency")
plt.show()
```

```python
if len(numeric_df.columns) >= 2:
    x = numeric_df.columns[0]
    y = numeric_df.columns[1]

    sns.scatterplot(x=x, y=y, data=df)
    plt.title("Multivariate Analysis")
    plt.show()
```
